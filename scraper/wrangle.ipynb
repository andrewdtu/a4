```python
import pandas as pd
```

```python
df = pd.read_csv('games-data_new.csv')
df
```

```python
df['r-date'] = pd.to_datetime(df['r-date'])
df['r-date'][0].year
```

```python
platform = {}
all_platforms ={}
for row in df.itertuples():
    if row.platform not in all_platforms:
        all_platforms[row.platform] = 0
for row in df.itertuples():
    #if row[0] == 0:
     #   print(row._3.year)
        
    
    if row._3.year not in platform:
        if len(platform) == 0:
            
            platform_year = {}
            platform[row._3.year] = all_platforms.copy()
            platform[row._3.year][row.platform] = 0
        elif len(platform) != 0:
            platform[row._3.year] = platform[row._3.year-1].copy()
    if row._3.year in platform:
        platform[row._3.year][row.platform] +=1
        #if row.platform not in platform_year:
         #   platform_year[row.platform] = 1
        #elif row.platform in platform_year:
         #   platform_year[row.platform] +=1
list_of_df = []
for outer_dict in platform:
    platform_series = pd.DataFrame.from_dict(platform[outer_dict],orient='index', columns =['count'])
    platform_series['year'] = outer_dict
    platform_series.reset_index(level=0, inplace=True)
    
    list_of_df.append(platform_series)
platform_df = pd.concat(list_of_df)
platform_df.reset_index(drop=True,inplace=True)
platform_df.rename(columns={'index':'platform'},inplace=True)
platform_df.to_csv('platform.csv')
platform_df
```

```python

all_genre = {}
df_listed_genre = df.copy()
df_listed_genre["genre_list"] = ''




for row in df.itertuples():
    if ',' in row.genre:
        df_listed_genre['genre_list'][row.Index] = row.genre.split(',') 
        for genre in df_listed_genre['genre_list'][row.Index]: 
            #print(df['genre'][row.Index])
            if genre not in all_genre:
                all_genre[genre] = 0
                
    elif ',' not in row.genre and type(row.genre) != list :
        df_listed_genre['genre_list'][row.Index] = row.genre.split(',') 
        for genre in df_listed_genre['genre_list'][row.Index]:    
            if genre not in all_genre:
                all_genre[genre] = 0
genre = {}

for row in df_listed_genre.itertuples():
    if row._3.year not in genre:
        if len(genre) == 0:
            genre_year = {}
            genre[row._3.year] = all_genre.copy()
            for i in row.genre_list:
                genre[row._3.year][i] = 0
        elif len(genre) != 0:
            genre[row._3.year] = genre[row._3.year-1].copy()
            
    if row._3.year in genre: 
        
        for i in row.genre_list: 
            if i in genre[row._3.year]:
                genre[row._3.year][i] +=1      
list_of_df = []
for outer_dict in genre:
    genre_series = pd.DataFrame.from_dict(genre[outer_dict],orient='index', columns =['count'])
    genre_series['year'] = outer_dict
    genre_series.reset_index(level=0, inplace=True)
    
    list_of_df.append(genre_series)
genre_df = pd.concat(list_of_df)
genre_df.reset_index(drop=True,inplace=True)
genre_df.rename(columns={'index':'genre'},inplace=True)
genre_df.to_csv('genre.csv')
genre_df
```
